### コードの実行

グラフの作成をMatplotlibを使って行います.  
Matplotlibは、Pythonのグラフ描画ライブラリです。  
折れ線グラフや棒グラフなど、様々なプロットを扱うことができます。  

```python
import matplotlib.pyplot as plt
import numpy as np

# データ生成
x = np.linspace(0, 10, 100)
y = x + np.random.randn(100) 

# プロット
plt.plot(x, y, label="test")

# 凡例の表示
plt.legend()

# プロット表示(設定の反映)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# データ生成
x = np.linspace(0, 10, 100)
y = x + np.random.randn(100)

# Figureの初期化
fig = plt.figure(figsize=(12, 8)) #...1

# Figure内にAxesを追加()
ax = fig.add_subplot(111)
ax.plot(x, y, label="test")
#ax2 = fig.add_subplot(111)
#ax.plot(y, x, label="test2")

# 凡例の表示
plt.legend()

# プロット表示(設定の反映)
plt.show()
```
